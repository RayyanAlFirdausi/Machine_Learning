```python
# 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

df_tugas = pd.read_csv('/content/drive/MyDrive/ML/wbc.csv')
df_tugas.info()
df_tugas.head()
```

```python
df_tugas = df_tugas.drop(['id', 'Unnamed: 32'], axis=1) # kode ini digunakan digunakan untuk menyingkirkan variabel yang tidak perlu
df_tugas.head()
```

```python
# 2. Lakukan proses encoding pada kolom "diagnosis".

from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder()

df_tugas['diagnosis'] = le.fit_transform(df_tugas['diagnosis'])
df_tugas.head()
```

```python
# 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

std = StandardScaler()

# agar lebih cepat, jadi menggunakan kode ini dengan memilih kolom bertipe data float64 dan int64
kolom_numerik = df_tugas.select_dtypes(include=['float64', 'int64']).columns
df_tugas[kolom_numerik] = std.fit_transform(df_tugas[kolom_numerik])
df_tugas.head()
```

```python
# 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

from sklearn.model_selection import train_test_split
df_tugas_train, df_tugas_test = train_test_split(df_tugas, test_size=0.2, random_state=0, stratify=df_tugas['diagnosis'])

print(f'Jumlah label data asli:\n{df_tugas.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_tugas_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_tugas_test.diagnosis.value_counts()}')
```
